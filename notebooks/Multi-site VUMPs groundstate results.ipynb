```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['text.usetex'] = True
plt.rcParams.update({'font.size': 14, 
                     'figure.figsize':(10, 6), 
                     'figure.facecolor': 'FFFFFF'})
```

```python
g_exact, e_exact = np.loadtxt('../data/exact_gs_energy.csv', delimiter=',')
n = 16

g_qb = np.linspace(0.1, 1.7, n)
```

## Ground state energy vs Hamiltonian site number

```python
e_4qb = np.load('../data/gstate_ising2_D2_qb4.npy', allow_pickle=True)
e_3qb = np.load('../data/gstate_ising2_D2_qb3.npy', allow_pickle=True)
e_2qb = np.load('../data/gstate_ising2_D2_qb2.npy', allow_pickle=True)
```

```python
plt.plot(g_exact, e_exact, label='exact')
plt.plot(g_qb, e_4qb, label='4 site')
plt.plot(g_qb, e_3qb, label='3 site')
plt.plot(g_qb, e_2qb, label='2 site')
plt.xlabel('g')
plt.ylabel(r'$<\psi|h_2|\psi>$')
plt.title('Ground state energy')
plt.legend()
plt.show()
```

```python
error2b = np.abs(e_2qb - e_exact)
error3qb = np.abs(e_3qb - e_exact)
error4qb = np.abs(e_4qb - e_exact)
```

```python
plt.plot(g_qb, error2b)
plt.title('Error 2 Site')
plt.xlabel('g')
plt.ylabel(r'$|E_{exact} - E_2|$')
```

```python
plt.plot(g_qb, error3qb)
plt.title('Error 3 Site')
plt.xlabel('g')
plt.ylabel(r'$|E_{exact} - E_3|$')
```

```python
plt.plot(g_qb, error4qb)
plt.title('Error 4 Site')
plt.xlabel('g')
plt.ylabel(r'$|E_{exact} - E_4|$')
```

```python
i = 4
plt.title('Ground state energy error')
plt.xlabel('g')
plt.ylabel(r'$|E_{exact} - E_n|$')
plt.plot(g_qb[:i], error4qb[:i], label='4 site')
plt.plot(g_qb[:i], error3qb[:i], label='3 site')
plt.plot(g_qb[:i], error2b[:i], label='2 site')
plt.legend()
```

### Ground state energy for different $D$ for $qNo = 2$

```python
e_D2 = np.load('../data/gstate_ising2_D2_qb2.npy', allow_pickle=True)
e_D4 = np.load('../data/gstate_ising2_D4_qb2.npy', allow_pickle=True)
e_D8 = np.load('../data/gstate_ising2_D8_qb2.npy', allow_pickle=True)
```

```python
error_D2 = np.abs(e_exact - e_D2)
error_D4 = np.abs(e_exact - e_D4)
error_D8 = np.abs(e_exact - e_D8)
```

```python
plt.semilogy(g_qb, error_D2, label='D = 2')
plt.semilogy(g_qb, error_D4, label="D = 4")
plt.semilogy(g_qb, error_D8, label="D = 8")
plt.xlabel('g')
plt.ylabel(r'$|E_{exact} - E_D|$')
plt.legend()
plt.title('Error 2 site, varying bond dimension')
plt.show()
```
